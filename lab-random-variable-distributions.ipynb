### Lab-Cleaning-Numerical-Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
customer_df=pd.read_csv("files_for_lab\we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python
# converting column names to snake type

customer_df.columns = [x.lower() for x in customer_df.columns]
customer_df.columns = customer_df.columns.str.replace(' ','_')
```

```python
# turning effective to date into datetime type

customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])

# checking for nulls

customer_df.isnull().sum() # - no nulls
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

numericals = customer_df.select_dtypes(include=numerics)

categoricals = customer_df.select_dtypes(include=object)
```

```python
def cont_disc(numericals):
    
    cont = pd.DataFrame()
    disc = pd.DataFrame()
    notdisc = cont.columns
    
    for col in numericals.columns:
        if numericals[col].dtypes == float:
            cont = pd.concat([cont,numericals[col]], axis=1)
     
    disc = numericals.drop(notdisc, axis=1)     
    
    return (cont,disc)   
```

```python
[continuous_df ,discrete_df ] = cont_disc(numericals)

display(continuous_df.head())
display(discrete_df .head())
```

```python
# The customer_id is a code unique to each customer, which would only add noise and no useful info to our analysis.
# Therefore the column should be dropped and disregarded - does not matter which customer is which.

categoricals = categoricals.drop('customer',axis=1)

categoricals.head()
```

```python
# Before plotting my numericals I'm going to normalize them so I can use my functions more effectively

# STANDARDSCALER

def stdnorm (df):
    
    standardnorm = StandardScaler().fit(df)
    x_st = standardnorm.transform(df)
    std_df = pd.DataFrame(x_st, columns=df.columns)
    
    return (std_df)
```

```python
std_continuous_df = stdnorm(continuous_df)
std_discrete_df = stdnorm(discrete_df)

display(std_continuous_df)
display(std_discrete_df)
```

```python
def disc_plot (disc):
    
    for col in disc.columns:
        sns.displot(disc[col], bins=20)
        plt.show()
```

```python
def cont_plot (cont):
    
    for col in cont.columns:
        plt.xlabel(col)
        plt.ylabel('Count')
        sns.kdeplot(cont[col],shade=True)
        plt.show()  
```

### Lab-Comparing-Regression-Models

```python
# 1st concatenate all my worked on dataframes: discrete + continuous + categorical + time

complete_df = pd.concat([std_discrete_df, std_continuous_df,One_Hot_Cat,time_df], axis=1)

# X-y Split

y = complete_df['total_claim_amount']
X = complete_df.drop(['total_claim_amount'], axis=1)

# Train-Test Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=32)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Running Regressions

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score

def reg_mod(X_train, X_test, y_train, y_test):
    
    # Run linear model
    lm = linear_model.LinearRegression()
    model = lm.fit(X_train,y_train)
    predictions = lm.predict(X_test)
    L_score = r2_score(y_test, predictions)
    # Run KNN model
    KNN = KNeighborsRegressor(n_neighbors=3)
    KNN.fit(X_train, y_train)
    KNN_score = KNN.score(X_test, y_test)
    # Run MLP model
    regr = MLPRegressor(random_state=1, max_iter=500).fit(X_train, y_train)
    MLP_score = regr.score(X_test, y_test)

    return (L_score, KNN_score, MLP_score)
```

```python
L_score, KNN_score, MLP_score = reg_mod(X_train, X_test, y_train, y_test)

print('Linear Reg. Score : ', L_score)
print('KNN Reg. Score : ', KNN_score)
print('MLP Reg. Score : ', MLP_score)
```
